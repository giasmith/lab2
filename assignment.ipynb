### b. Use requests and BeautifulSoup to save the list of song lyric URLs in a Python list.

```python
r = requests.get(urltoscrape, 
                 headers= {'User-Agent': useragent,
                           'From': 'nqj5zk@virginia.edu'})
lizzohtml=BeautifulSoup(r.text, 'html.parser')
```

```python
song_urls = []
p2 = lizzohtml.find_all('ol')[0]

for link in p2.find_all('a', attrs={'href': re.compile("https://")}):
    song_urls.append(link.get('href'))
    
song_urls
```

### c. Build a spider

```python
def time_2_scrape(song_urls):
    time.sleep(2.5)
    print("Scraping the queen's song:" + song_urls)
    r = requests.get(song_urls, headers= {'User-Agent': useragent,
                                            'From': 'nqj5zk@virginia.edu'})
    lizzohtml=BeautifulSoup(r.text, 'html.parser')
    lizzo_txt = lizzohtml.find_all('div', id = 'lyrics-root-pin-spacer')[0].text
    return lizzo_txt
```
